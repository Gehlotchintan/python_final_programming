```python
# case study 2
#1 Python program to find sum of even factor of a number
def evenfactorssum(n):
    s = 0
    for i in range(1, n+1):
        # finding factors of a given number
        if n % i == 0:
            # finding even factors of a given number
            if i % 2 == 0:
                # adding even factors of a given number
                s = s+i  # 2+6+10+30
                # printing the sum of even factors of a given number
    print(s)



evenfactorssum(30)

```

```python
# python program for smallest k digit number divisible by x
def small_k(n, d):

   Min = pow(10, d-1)
   if(Min%n == 0):
      return (Min)
   else:
      return ((Min + n) - ((Min + n) % n))

small_k(83, 5)

```

```python
# 3 python program for most occuring character and its count
def most_occ(x):
    fre_count={}
    for i in x:
        if i in fre_count:
            fre_count[i]=fre_count[i]+1
        else:
            fre_count[i]=1
    return (max(fre_count,key=fre_count.get) , fre_count[max(fre_count,key=fre_count.get)])
most_occ("hello")


```
