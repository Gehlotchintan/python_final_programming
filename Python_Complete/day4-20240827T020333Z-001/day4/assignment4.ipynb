# Fibonacci series
-   input: 6
-   output: 0 1 1 2 3 5

```python
FibArray = [0, 1]

def fibonacci(n):

    # Check is n is less
    # than 0
    if n < 0:
        print("Incorrect input")

    # Check is n is less
    # than len(FibArray)
    elif n < len(FibArray):
        return FibArray[n]
    else:
        FibArray.append(fibonacci(n - 1) + fibonacci(n - 2))
    return FibArray

# Driver Program
print(fibonacci(9))
```

# Factorial of number
-   input: 5
-   output: 120

```python
def factorials (n):
  f=1
  for i in range(1,n+1):
    f=f*i
  return f
factorials(5)

```

# Armstrong number
-   input: 153
-   output: Armstrong Number
-   Explanation: 1^3 + 5^3 + 3^3 == 153,   why power of 3, coz of 3 digit

```python
n = 153  # or n=int(input()) -> taking input from user
s = n  # assigning input value to the s variable
b = len(str(n))
sum1 = 0
while n != 0:
    r = n % 10
    sum1 = sum1+(r**b)
    n = n//10
if s == sum1:
    print("The given number", s, "is armstrong number")
else:
    print("The given number", s, "is not armstrong number")
```

# Palindrome of a String
-   input : aba
-   output: Palindrome
-   Explanation: reverese of "aba" is equals to "aba"


```python
def pelindrome(m):
  if m==m[::-1]:
    print('given string is palindrome')
  else:
    print('not a palindrome')
pelindrome('papa')
```

```python
def pelindrome(m):
  rev=''
  for i in range(1,len(m)+1):
    rev=rev+m[-i]
  if m==rev:
    print('given string is palindrome')
  else:
    print('not a palindrome')
pelindrome('sas')
```

# Palindrome of a Number
