# Function with parameters/args

```python
def isEven(num: int) -> None:
    if num % 2 == 0:
        print("Even")
    else:
        print("Odd")
```

```python
isEven(4)
```

```python
def isDivisible(num: int, divisor: int) -> None:
    if num % divisor == 0:
        print("Divisible")
    else:
        print("Not divisible")
```

```python
isDivisible(10, 2)
```

# Dynamic Args/kwargs

```python
def square(*nums):
    sq = []
    for num in nums:
        sq.append(num**2)
    return sq
```

```python
square(2, 3,4,5,6)
```

```python
def getInfo(**kwargs):
    for key, val in kwargs.items():
        print(key, val)
```

```python
getInfo(name='Alle', email='alley@g.com', age=24)
```

```python
# Global scope
num = 8


def power():
    num = 9     # local scope
    print(num**2)
```

```python
power()
```

```python
# Global scope
num = 8


def power():
    num = 8**2  # local scope
```

```python
power()
```

```python
num
```

```python
# Global scope
num = 8


def power():
    global num
    num = 8**2
```

```python
power()
```

```python
num
```

```python
# Enclosed scope
num = 8


def power():
    global num
    num = 8**2
    def multiply():
        print(num*2)
    multiply()
    print(num)
```

```python
power()
```

# lambda
<img src="lambda.jpg">

```python
def isDivisible(num: int = 1, divisor: int = 1) -> bool:
    if num % divisor == 0:
        return True
    return False
```

```python
check = lambda num, divisor: True if num%divisor==0 else False
check(10, 2)
```

```python
genPowers2 = lambda start, stop: [num**2 for num in range(start, stop)]
genPowers2(2, 8)
```

```python
# y = mx + c
y = lambda m, x, c: m*x + c
y(2, 3, 1)
```
