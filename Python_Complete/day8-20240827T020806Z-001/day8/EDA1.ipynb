# Joins

```python
humidity = pd.read_csv('humidity.csv')
temp = pd.read_csv('temperature.csv')
```

```python
humidity
```

```python
temp
```

```python
# inner join
```

```python
temp.merge(humidity, on='city', how='inner')
```

```python
# left join
humidity.merge(temp, on='city', how='left')
```

```python
# right join
humidity.merge(temp, on='city', how='right')
```

```python
# cross join
humidity.merge(temp,how='cross')
```
