# String Formatting

```python
name = "Alley"
a = 24
w = 67.78
```

```python
print("Name is ", name, ". Age is ", a, ". Weight is ", w)
```

```python
print("Name is {0}. Age is {1}. Weight is {2}".format(name, a, w))
```

```python
print("Name is %s. Age is %d. Weight is %.2f" %(name, a, w))
```

```python
print(f"Name is {name} Age is {a}. Weight is {w: .1f}")
```
