# shallow vs deep copy

```python
# shallow copy
data1 = [2,3,4]
copy = data1
copy.append(99)
copy
```

```python
data1
```

```python
# deep copy
data1 = [2,3,4]
copy = data1.copy()
copy.append(88)
copy
```
